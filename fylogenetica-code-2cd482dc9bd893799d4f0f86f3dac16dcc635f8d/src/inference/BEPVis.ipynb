```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f = open("C:/Users/20202991/Dropbox/My PC (S20202991)/Desktop/Inference_Analysis3.txt")
content = f.read()
content = content.replace('[', '')
content = content.replace(']', '')
content = content.replace(',', '')
content = content.replace('\n', '_')
parts = content.split("_")
parts2 = [part.split(" ") for part in parts]
numbers = []
for part in parts2:
    partial = []
    for num in part:
        partial.append(int(num))
    numbers.append(partial)
df = pd.DataFrame(numbers, columns=['taxa','1','2','3','4','5'])
print(df.head())
#df2.sort_values('taxa')
df2 = df.groupby('taxa')[['1','2','3','4','5']].mean()
df3 = df.groupby('taxa')[['1','2','3','4','5']].min()
df4 = df.groupby('taxa')[['1','2','3','4','5']].max()
df2['1'] = df2['1']/(df2.index*df2.index)#*(df2.index-1)*(df2.index-2)*(df2.index-3))
df2['2'] = df2['2']/(df2.index*df2.index)#*(df2.index-1)*(df2.index-2)*(df2.index-3))
df2['3'] = df2['3']/(df2.index*df2.index)#*(df2.index-1)*(df2.index-2)*(df2.index-3))
df2['4'] = df2['4']/(df2.index*df2.index)#*(df2.index-1)*(df2.index-2)*(df2.index-3))
df2['5'] = df2['5']/(df2.index*df2.index)#*(df2.index-1)*(df2.index-2)*(df2.index-3))
print(df2)
print(df3)
print(df4)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df2['1'], label="no inference", color="b", linestyle="-")
ax1.plot(df2['2'], label="2 inference", color="g", linestyle="-")
ax1.plot(df2['3'], label="3 inference", color="r", linestyle="-")
ax1.plot(df2['4'], label="4 inference", color="c", linestyle="-")
ax1.plot(df2['5'], label="5 inference", color="m", linestyle="-")
#ax1.plot(df3['1'], label="1 min", color="b", linestyle="--")
#ax1.plot(df3['2'], label="2 min", color="g", linestyle="--")
#ax1.plot(df3['3'], label="3 min", color="r", linestyle="--")
#ax1.plot(df3['4'], label="4 min", color="c", linestyle="--")
#ax1.plot(df3['5'], label="5 min", color="m", linestyle="--")
#ax1.plot(df4['1'], label="1 max", color="b", linestyle=":")
#ax1.plot(df4['2'], label="2 max", color="g", linestyle=":")
#ax1.plot(df4['3'], label="3 max", color="r", linestyle=":")
#ax1.plot(df4['4'], label="4 max", color="c", linestyle=":")
#ax1.plot(df4['5'], label="5 max", color="m", linestyle=":")
plt.xlabel("Number of taxa")
plt.ylabel("Number of quartets/(Number of taxa)^2")
ax1.legend()
plt.show()
```

```python
df2['div2'] = df2['2']/df2['1']*100
df2['div3'] = df2['3']/df2['1']*100
df2['div4'] = df2['4']/df2['1']*100
df2['div5'] = df2['5']/df2['1']*100
print(df2)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df2['div2'], label="2 inference")
ax1.plot(df2['div3'], label="3 inference")
ax1.plot(df2['div4'], label="4 inference")
ax1.plot(df2['div5'], label="5 inference")
ax1.legend()
plt.ylabel("Percentage")
plt.xlabel("Number of taxa")
plt.ylim([50,100])
plt.show()
```
